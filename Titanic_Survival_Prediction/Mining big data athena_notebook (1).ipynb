### Reading saved parquet data 

```python_session
%%sql
CREATE EXTERNAL TABLE IF NOT EXISTS titanic_data (
    PassengerId INT,
    Survived INT,
    Pclass INT,
    Name STRING,
    Sex STRING,
    Age DOUBLE,
    SibSp INT,
    Parch INT,
    Ticket STRING,
    Fare DOUBLE,
    Embarked STRING
)
STORED AS PARQUET
LOCATION 's3://titanic1m/titanic_data/'
```

```python_session
%%sql
SELECT * FROM titanic_data LIMIT 10
```

```python_session
%%sql
SELECT Sex, AVG(Age) as AvgAge, AVG(Fare) as AvgFare
FROM titanic_data
GROUP BY Sex
```

```python_session
%%sql
SELECT COUNT(*) as TotalRows FROM titanic_data
```

```python_session
%%sql
SELECT
    MIN(Age) AS min_age, MAX(Age) AS max_age, AVG(Age) AS avg_age,
    MIN(Fare) AS min_fare, MAX(Fare) AS max_fare, AVG(Fare) AS avg_fare,
    MIN(SibSp) AS min_sibsp, MAX(SibSp) AS max_sibsp, AVG(SibSp) AS avg_sibsp,
    MIN(Parch) AS min_parch, MAX(Parch) AS max_parch, AVG(Parch) AS avg_parch
FROM titanic_data
```

```python_session
%%sql
SELECT Sex, COUNT(*) AS count
FROM titanic_data
GROUP BY Sex
```

```python_session
%%sql
SELECT CAST(Pclass AS STRING), COUNT(*) AS count
FROM titanic_data
GROUP BY Pclass
```

```python_session
%%sql
SELECT Embarked, COUNT(*) AS count
FROM titanic_data
GROUP BY Embarked
```

```python_session
%%sql
SELECT CAST(Survived AS STRING), COUNT(*) AS count
FROM titanic_data
GROUP BY Survived
```

```python_session
%%sql
SELECT 
    AVG(Age) AS avg_age,
    MIN(Age) AS min_age,
    MAX(Age) AS max_age,
    AVG(Fare) AS avg_fare,
    MIN(Fare) AS min_fare,
    MAX(Fare) AS max_fare,
    AVG(SibSp) AS avg_sibsp,
    MAX(SibSp) AS max_sibsp,
    AVG(Parch) AS avg_parch,
    MAX(Parch) AS max_parch
FROM titanic_data
```

```python_session
%%sql
SELECT 
    CASE 
        WHEN Age < 18 THEN 'Child'
        WHEN Age BETWEEN 18 AND 30 THEN 'Young Adult'
        WHEN Age BETWEEN 31 AND 50 THEN 'Adult'
        WHEN Age > 50 THEN 'Senior'
        ELSE 'Unknown'
    END AS age_group,
    COUNT(*) AS count
FROM titanic_data
GROUP BY age_group
ORDER BY count DESC
```

```python_session
%%sql
SELECT 
    Pclass,
    AVG(CAST(Survived AS DOUBLE)) AS survival_rate,
    COUNT(*) AS total_passengers
FROM titanic_data
GROUP BY Pclass
ORDER BY Pclass
```

```python_session
%%sql
SELECT 
    Sex,
    AVG(CAST(Survived AS DOUBLE)) AS survival_rate,
    COUNT(*) AS total_passengers
FROM titanic_data

GROUP BY Sex
```

```python_session
%%sql
SELECT 
    Pclass,
    Survived,
    AVG(Fare) AS avg_fare
FROM titanic_data
GROUP BY Pclass, Survived
ORDER BY Pclass, Survived
```

```python_session
%%sql
SELECT 
    (SibSp + Parch + 1) AS family_size,
    COUNT(*) AS count
FROM titanic_data
GROUP BY family_size
ORDER BY family_size
```

```python_session
%%sql
SELECT 
    CASE 
        WHEN Age < 18 THEN 'Child'
        WHEN Age BETWEEN 18 AND 30 THEN 'Young Adult'
        WHEN Age BETWEEN 31 AND 50 THEN 'Adult'
        WHEN Age > 50 THEN 'Senior'
        ELSE 'Unknown'
    END AS age_group,
    Sex,
    AVG(CAST(Survived AS DOUBLE)) AS survival_rate,
    COUNT(*) AS total_passengers
FROM titanic_data
GROUP BY age_group, Sex
ORDER BY age_group, Sex
```

```python_session
%%sql
SELECT 
    CASE 
        WHEN (SibSp + Parch) = 0 THEN 'Alone'
        WHEN (SibSp + Parch) BETWEEN 1 AND 3 THEN 'Small Family'
        WHEN (SibSp + Parch) BETWEEN 4 AND 6 THEN 'Medium Family'
        ELSE 'Large Family'
    END AS family_size_category,
    AVG(CAST(Survived AS DOUBLE)) AS survival_rate,
    COUNT(*) AS total_passengers
FROM titanic_data
GROUP BY family_size_category
ORDER BY survival_rate DESC
```

```python_session
%%sql
SELECT 
    Pclass,
    Embarked,
    AVG(CAST(Survived AS DOUBLE)) AS survival_rate,
    COUNT(*) AS total_passengers
FROM titanic_data
WHERE Embarked IS NOT NULL
GROUP BY Pclass, Embarked
ORDER BY Pclass, survival_rate DESC
```

```python_session
%%sql
SELECT 
    Pclass,
    Survived,
    AVG(Fare) AS avg_fare,
    COUNT(*) AS total_passengers
FROM titanic_data
GROUP BY Pclass, Survived
ORDER BY Pclass, Survived
```

```python_session
%%sql
WITH fare_quartiles AS (
    SELECT 
        Fare,
        Survived,
        NTILE(4) OVER (ORDER BY Fare) AS quartile
    FROM titanic_data
    WHERE Fare > 0
)
SELECT 
    quartile,
    MIN(Fare) AS min_fare,
    MAX(Fare) AS max_fare,
    AVG(CAST(Survived AS DOUBLE)) AS survival_rate,
    COUNT(*) AS total_passengers
FROM fare_quartiles
GROUP BY quartile
ORDER BY quartile
```
